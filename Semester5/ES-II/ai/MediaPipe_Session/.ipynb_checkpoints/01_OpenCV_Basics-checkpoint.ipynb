Import the required libraries 

```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
originalImg = cv.imread('./data/Champions.jpg')
#cv.imshow('Champions 2023', originalImg)
#cv.waitKey(0)
#cv.destroyAllWindows()
```

```python
def displayImage(path):
    img = mpimg.imread(path)
    imgplot = plt.imshow(img)
    plt.show()
```

```python
displayImage('./data/Champions.jpg')
```

Change the image channel

```python
grayImg = cv.cvtColor(originalImg, cv.COLOR_BGR2GRAY)
cv.imwrite('./data/grayscale.jpg',grayImg)
#displayImage('./data/grayscale.jpg')
img = mpimg.imread('./data/grayscale.jpg')
#plt.axis("off")
imgplot = plt.imshow(img, cmap = 'gray')
plt.show()
```

Image resize

```python
originalImg = cv.imread('./data/Champions.jpg')
print(originalImg.shape)
```

```python
dim = (600,400)
resizedImg = cv.resize(originalImg, dim, interpolation = cv.INTER_AREA)
cv.imwrite('./data/resized_Champions.jpg', resizedImg)
displayImage('./data/resized_Champions.jpg')
print(resizedImg.shape)
```

Image Filtering

```python
originalImg = cv.imread('./data/Champions.jpg')
```

```python
kernel_size = (25, 25)
gBlur = cv.GaussianBlur(originalImg, kernel_size, 0)
cv.imwrite('./data/gBlur_Champions.jpg', gBlur)
displayImage('./data/gBlur_Champions.jpg')
```

```python
mBlur = cv.medianBlur(originalImg, 25)
cv.imwrite('./data/mBlur_Champions.jpg', mBlur)
displayImage('./data/mBlur_Champions.jpg')
```

Live camera streaming 

```python
video = cv.VideoCapture(0)
while(True):
    _,frame = video.read()
    cv.imshow('Webcam', frame)
    if cv.waitKey(1) & 0xFF == ord('q'):
        break

video.release()
```
