```python
import cv2
import mediapipe as mp

# Initialize MediaPipe Holistic
mp_holistic = mp.solutions.holistic
holistic = mp_holistic.Holistic(
    min_detection_confidence=0.5, 
    min_tracking_confidence=0.5,
    model_complexity=2 # Optional: Increases processing time but more accurate
)

# Start video capture
cap = cv2.VideoCapture(1)

while cap.isOpened():
    ret, frame = cap.read()
    frame = cv2.flip(frame, 1)
    if not ret:
        break

    try:
        # Convert the frame to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = holistic.process(image)

        # Convert back to BGR for OpenCV
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)

        # Draw landmarks if they exist
        if results.face_landmarks:
            mp.solutions.drawing_utils.draw_landmarks(
                image, results.face_landmarks, mp_holistic.FACE_CONNECTIONS)
        else:
            print("No face landmarks detected, continuing to check...")

        if results.pose_landmarks:
            mp.solutions.drawing_utils.draw_landmarks(
                image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS)
        else:
            print("No pose landmarks detected, continuing to check...")

    except Exception as e:
        print(f"Error processing frame: {e}")

    # Display the frame
    cv2.imshow('MediaPipe Holistic', image)

    # Exit the loop when 'q' is pressed
    if cv2.waitKey(5) & 0xFF == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```
